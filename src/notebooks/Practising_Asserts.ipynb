## Sample Code

```python
from pyspark.sql import functions as F

data = [
    ("2025-09-01 08:00:00", "2025-09-01 08:30:00", 2.5, 15.0, 10001, 10005),
    ("2025-09-01 09:00:00", "2025-09-01 09:45:00", 5.2, 25.0, 10150, 10281),
    ("2025-09-01 10:00:00", "2025-09-01 10:20:00", 1.8, 12.0, 10300, 10003),
]

columns = ["pickup_datetime", "dropoff_datetime", "trip_distance", "fare_amount", "pickup_zip", "dropoff_zip"]

df = spark.createDataFrame(data, columns)
```

```python
df.show()
```

```python
def add_city(field):
    if field > 10001 and field < 10006:
        return "Manhattan"
    else:
        return "New York"

city_udf = F.udf(add_city, "string")

df = df.withColumn("pickup_city", city_udf(df["pickup_zip"]))

df.display()
```
