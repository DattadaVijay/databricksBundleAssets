```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import mlflow
import mlflow.sklearn
from mlflow.models.signature import infer_signature
import pandas as pd

# Load data
X, y = load_iris(return_X_y=True)
X_df = pd.DataFrame(X, columns=["sepal_length","sepal_width","petal_length","petal_width"])
y_df = pd.DataFrame(y, columns=["species"])

# Train model
clf = RandomForestClassifier(n_estimators=10, random_state=42)
clf.fit(X_df, y_df)

# Infer signature
signature = infer_signature(X_df, clf.predict(X_df))

# Log & register model with signature
with mlflow.start_run():
    mlflow.sklearn.log_model(
        sk_model=clf,
        artifact_path="iris_rf_model",
        registered_model_name="iris_rf_classifier",
        signature=signature
    )
```
